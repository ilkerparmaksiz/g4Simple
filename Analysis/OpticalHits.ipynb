```python
import uproot as up
import matplotlib.pyplot as plt
```

```python
f1=up.open("../build/result_t0.root")
```

```python
hits=f1["ohits"]
```

```python
hits.keys()
```

```python
print(hits["det_name"].array().to_numpy())
```

```python
evtid=hits["evtid"].array().to_numpy() # Mev to ev
#print(evtid)
```

```python
energy=hits["edep"].array().to_numpy()*1e6 # MeV to ev 
print(len(energy))
print(f"Max Energy {round(energy.max(),2)} eV")
print(f"Min Energy {round(energy.min(),2)} eV")
```

```python
_=plt.hist(energy,bins=40)
plt.xlabel("Energy (eV)")
```

```python
x=hits["x"].array().to_numpy()
y=hits["y"].array().to_numpy()
_=plt.hist2d(x,y,bins=100)
plt.colorbar(label="photons")
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
```
